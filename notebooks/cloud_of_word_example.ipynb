```python
import pandas as pd
from text_tools import preprocessing

df=pd.read_csv("/Users/leffepierre/Documents/Ynov/ML/teaching_ML_project/en.openfoodfacts.org.products.csv",
                     sep="\t", encoding="utf-8", nrows=100000, low_memory=False)
df=preprocessing(df,"categories_en")
df
```

```python
from wordcloud import WordCloud
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = ["food","and"], 
                min_font_size = 10).generate(' '.join(df['categories_en']))

import matplotlib.pyplot as plt
  
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```
